```python
k=geo_datadf.groupby('Municipio').size().reset_index(name='Frecuencia')

# Paso 2: Combinar los datos con beat_data
k = geo_datadf.merge(k, how='left', left_on='Municipio', right_on='Municipio')
#k = k.dropna(subset=['geometry'])
k.Frecuencia[k.Frecuencia>1000] = 0
```

```python
columns_to_include = ['ETC', 'MPIO_CNMBR','Frecuencia','geometry']
k_temp = k[columns_to_include]
k_temp = k_temp.set_index('MPIO_CNMBR')
k_temp
```
